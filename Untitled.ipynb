## https://www.linkedin.com/in/shahrour/details/skills/

```python
def get_skills(filename:str):
    page_soup = prepaer_file_to_scrap(filename)
    skills_all = []
    skills = page_soup.findAll("span", {"class": "mr1 hoverable-link-text t-bold"})
    for item in skills:
        x = len(item.text)/2
        skills_all.append(item.text[:int(x)])
    return skills_all

get_skills('(2) Skills _ Abdalrahman Shahrour _ LinkedIn.html')
```

## https://www.linkedin.com/in/shahrour/details/experience/

```python
def experience(filename:str):
    page_soup = prepaer_file_to_scrap(filename)
    comp, pos1, time = [], [], []
    edu = page_soup.findAll("span", {"class": "t-14 t-normal"})
    for i in edu:
        comp.append(i.text[:int(len(i.text)/2)])
    pos = page_soup.findAll("span", {"class": "mr1 t-bold"})
    for i in pos:
        pos1.append(i.text[:int(len(i.text)/2)])
    timee = page_soup.findAll("span", {"class": "t-14 t-normal t-black--light"})
    for i in timee:
        pos1.append(i.text[:int(len(i.text)/2)])
    return comp, pos1, time

experience('(2) Experience _ Abdalrahman Shahrour _ LinkedIn.html') # t-14 t-normal t-black--light
```
